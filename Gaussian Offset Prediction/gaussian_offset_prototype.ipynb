```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
def generate_offsets(axis, n = 50, sigma = 0.1):
    offsets = np.random.normal(loc = 0, scale = sigma, size = (n, 6))
    offsets[:, 5] = 0
    np.savetxt("offsets_" + axis + ".csv", offsets, delimiter = ",")
    return offsets
```

```python
def generate_test_offsets(axis):
    offsets = np.zeros((12, 6))
    if axis == "y":
        for i in range(0, 6):
            offsets[i,i] = 0.2
    else:
        for i in range(6, 12):
            offsets[i,i-6] = 0.2
    np.savetxt("offsets_test_" + axis + ".csv", offsets, delimiter = ",")
    return offsets
```

```python
generate_test_offsets("y")
generate_test_offsets("z")
```

```python
offsets_y = generate_offsets(axis = "y", n = 50, sigma = 0.1)
offsets_z = generate_offsets(axis = "z", n = 50, sigma = 0.1)
```
